The purpose of this file is to reformat the course data extracted from the "Data Preparation" file, so that we can input the data into the traditional machine learning models.

```python
import numpy as np
import pandas as pd
```

```python
# Import course data from the "Data Preparation" file

BBB_Train = pd.read_csv("BBB_Train_df.csv", index_col = 0)
DDD_Train = pd.read_csv("DDD_Train_df.csv", index_col = 0)
FFF_Train = pd.read_csv("FFF_Train_df.csv", index_col = 0)

BBB_Test = pd.read_csv("BBB_Test_df.csv", index_col = 0)
DDD_Test = pd.read_csv("DDD_Test_df.csv", index_col = 0)
FFF_Test = pd.read_csv("FFF_Test_df.csv", index_col = 0)
```

```python
BBB_Train
```

```python
# Transform the "-1" values to "0" because the traditional models cannot make use of the padding functionality
BBB_Train.loc[(BBB_Train['sum_click'] == -1), 'sum_click':'Avg_score'] = 0
DDD_Train.loc[(DDD_Train['sum_click'] == -1), 'sum_click':'Avg_score'] = 0
FFF_Train.loc[(FFF_Train['sum_click'] == -1), 'sum_click':'Avg_score'] = 0

BBB_Test.loc[(BBB_Test['sum_click'] == -1), 'sum_click':'Avg_score'] = 0
DDD_Test.loc[(DDD_Test['sum_click'] == -1), 'sum_click':'Avg_score'] = 0
FFF_Test.loc[(FFF_Test['sum_click'] == -1), 'sum_click':'Avg_score'] = 0
```

```python
# Make the target label binary (0 for "fail", 1 for "pass" and "distinction")
BBB_Train["final_result"][BBB_Train["final_result"] == "Distinction"] = "Pass"
BBB_Train["final_result"][BBB_Train["final_result"] == "Pass"] = 1
BBB_Train["final_result"][BBB_Train["final_result"] == "Fail"] = 0
DDD_Train["final_result"][DDD_Train["final_result"] == "Distinction"] = "Pass"
DDD_Train["final_result"][DDD_Train["final_result"] == "Pass"] = 1
DDD_Train["final_result"][DDD_Train["final_result"] == "Fail"] = 0
FFF_Train["final_result"][FFF_Train["final_result"] == "Distinction"] = "Pass"
FFF_Train["final_result"][FFF_Train["final_result"] == "Pass"] = 1
FFF_Train["final_result"][FFF_Train["final_result"] == "Fail"] = 0

BBB_Test["final_result"][BBB_Test["final_result"] == "Distinction"] = "Pass"
BBB_Test["final_result"][BBB_Test["final_result"] == "Pass"] = 1
BBB_Test["final_result"][BBB_Test["final_result"] == "Fail"] = 0
DDD_Test["final_result"][DDD_Test["final_result"] == "Distinction"] = "Pass"
DDD_Test["final_result"][DDD_Test["final_result"] == "Pass"] = 1
DDD_Test["final_result"][DDD_Test["final_result"] == "Fail"] = 0
FFF_Test["final_result"][FFF_Test["final_result"] == "Distinction"] = "Pass"
FFF_Test["final_result"][FFF_Test["final_result"] == "Pass"] = 1
FFF_Test["final_result"][FFF_Test["final_result"] == "Fail"] = 0
```

Reformat dataframes into 2D arrays where the sequential dimension is combined with the feature dimension

```python
BBB_Train_pt = (BBB_Train.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())
DDD_Train_pt = (DDD_Train.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())
FFF_Train_pt = (FFF_Train.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())

BBB_Test_pt = (BBB_Test.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())
DDD_Test_pt = (DDD_Test.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())
FFF_Test_pt = (FFF_Test.pivot_table(index=['code_presentation', 'identifier', 'final_result'], columns='date', values=['sum_click', 'csum_asgmts', 'Avg_score']).reset_index())
```

```python
BBB_Train_pt.columns = BBB_Train_pt.columns.map('{0[0]}|{0[1]}'.format)
BBB_Train_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)
DDD_Train_pt.columns = DDD_Train_pt.columns.map('{0[0]}|{0[1]}'.format)
DDD_Train_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)
FFF_Train_pt.columns = FFF_Train_pt.columns.map('{0[0]}|{0[1]}'.format)
FFF_Train_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)

BBB_Test_pt.columns = BBB_Test_pt.columns.map('{0[0]}|{0[1]}'.format)
BBB_Test_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)
DDD_Test_pt.columns = DDD_Test_pt.columns.map('{0[0]}|{0[1]}'.format)
DDD_Test_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)
FFF_Test_pt.columns = FFF_Test_pt.columns.map('{0[0]}|{0[1]}'.format)
FFF_Test_pt.rename(columns={'code_presentation|':'code_presentation', 'identifier|':'identifier', 'final_result|':'final_result'}, inplace=True)
```
